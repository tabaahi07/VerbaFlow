```python
with open('Dataset_messege.txt', 'r') as file:
    data = file.read()
print(data)
```

```python
len(data)
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
tokenizer = Tokenizer()
tokenizer.fit_on_texts([data])
tokenizer.word_index
```

```python
total_words = len(tokenizer.word_index)
total_words
```

```python
input_sequence = []
for sen in data.split('\n') :
   tokenized_seq = tokenizer.texts_to_sequences([sen])[0]
   #print(tokenized_seq)
   for i in range(1,len(tokenized_seq)) :
     input_sequence.append(tokenized_seq[:i+1])

input_sequence
```

```python
max_len = max(len(seq) for seq in input_sequence)
max_len
```

```python
from keras.utils import pad_sequences
padded_input_sequences = pad_sequences(
    input_sequence ,
    maxlen = max_len ,
    padding = 'pre'
)
padded_input_sequences
```

```python
X = padded_input_sequences[:, :-1]
Y = padded_input_sequences[:,-1]
X
```

```python
Y = keras.utils.to_categorical(
    Y , num_classes = total_words
)
Y
```

```python
# Model Building
from keras.models import Sequential
from keras.layers import Embedding , LSTM , Dense

model = Sequential()
# 870=total no. of words in vocab
# Now each word was represnted by a vector after ohe , but now using embedding we are making it more dense and will have a size of 100
# So here total weights will be 870 * 100
model.add(Embedding(870,100,input_length=11))
# Only mention no. of nodes in LSTM
model.add(LSTM(100))
# In dense we'll add as many nodes as the no. of words in vocab to get prob of each word
model.add(Dense(870,activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy' , optimizer='adam' , metrics = ['accuracy'])
model.summary()
# Now now basically we have some input sequences and each sequence has some words which are represented by a vector of size 100 after embedding
# Now , in LSTM each input sequence is feeded in time steps i.e its first word's vector will be feede at t=0 n so on
# Now , after the last word feeded from an input seq , we will feed the output to dense layer which will give an output of sie 283 ,
# having prob of all the words and one which has highest prob will be treated as predicted output
```

```python
model.fit(X,Y,epochs=100)
```

```python
# Predicting
import numpy as np
text = "ghumne"

for i in range(10):
  # tokenize
  tokenized_text = tokenizer.texts_to_sequences([text])[0]
  # padding
  padded_tokenized_text = pad_sequences([tokenized_text], maxlen=11, padding='pre')
  # predict
  pos = np.argmax(model.predict(padded_tokenized_text))

  for word,index in tokenizer.word_index.items():
    if index == pos:
      text = text + " " + word
      print(text)
```
